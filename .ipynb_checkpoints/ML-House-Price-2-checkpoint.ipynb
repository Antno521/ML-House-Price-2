```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv('House Price Prediction Dataset.csv')
```

```python
# EDA
```

```python
# Looking at the data
print(dataset.head(5))
dataset.shape
print(dataset.dtypes)
```

```python
# Interval of the numerical features
print("Area:")
print("max: ", max(dataset['Area']))
print("min: ", min(dataset['Area']))

print("Bedrooms:")
print("max: ", max(dataset['Bedrooms']))
print("min: ", min(dataset['Bedrooms']))

print("Bathrooms:")
print("max: ", max(dataset['Bathrooms']))
print("min: ", min(dataset['Bathrooms']))

print("Floors:")
print("max: ", max(dataset['Floors']))
print("min: ", min(dataset['Floors']))

print("YearBuilt:")
print("max: ", max(dataset['YearBuilt']))
print("min: ", min(dataset['YearBuilt']))

print("Price:")
print("max: ", max(dataset['Price']))
print("min: ", min(dataset['Price']))
```

```python
# Heatmap to observe correlation between numerical features
numerical_dataset = dataset.select_dtypes(include=['number'])

plt.figure(figsize=(12,6))
sns.heatmap(numerical_dataset.corr(), cmap='BrBG', fmt='.2f', linewidths=2, annot=True)
```

```python
# Side by side scatter plots of the correlation between the numerical features
sns.pairplot(numerical_dataset)
```
